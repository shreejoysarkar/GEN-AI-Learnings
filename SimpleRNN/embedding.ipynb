```python
import tensorflow
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## Sentences

sent = [
    'I love my dog',
    'I love my cat',
    'I am a good boy',
    'I am a good developer',
    'I love to code',
    'I love to eat biryani',
]
```

```python
sent
```

```python
## Define the vocabulary size
vocab_size = 10000
```

```python
## One-hot representation the sentences
one_hot_sentences = [one_hot(words, vocab_size) for words in sent]
one_hot_sentences
```

```python
## Word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(sequences = one_hot_sentences , padding = 'pre', maxlen = sent_length)  
# post padding for adding zeros at end, making the length same
```

```python
embedded_docs
```

```python
## feature reqpresentation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(input_dim = vocab_size, output_dim = dim))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
