## Bag of words implementation

```python
import pandas as pd
```

```python
spam_msgs = pd.read_csv('spam.csv', encoding='latin-1', usecols=['v1', 'v2'])
```

```python
spam_msgs.head()
## changing the name of the columns
spam_msgs.columns = ['label', 'message']
```

```python
spam_msgs
```

```python
## Data cleaning and preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(spam_msgs)):
    review = re.sub('[^a-zA-Z]',' ',spam_msgs['message'][i])
    review = review.lower()
    review = review.split()
    # print(review)
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    # print(review)
    corpus.append(review)
```

```python
corpus
```

```python
### Creating the Bag of Words model

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=2500, binary=True) ## For binart BOW , we use binary=True
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
print(X)
```

```python
cv.vocabulary_
```

# N-grams

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=500, binary=True, ngram_range=(2,2))
X = cv.fit_transform(corpus).toarray()
```
