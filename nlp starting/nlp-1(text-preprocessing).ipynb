```python
corpus = '''Hellow world. This is a sample corpus for Natural Language Processing tasks. NLP is fun and exciting!
We will explore various techniques in NLP using Python. Let's get started!'''
```

```python
## Tokenization

import nltk
from nltk.tokenize import sent_tokenize
nltk.download('punkt_tab')
sentences = sent_tokenize(corpus)       ### sentence  -- > Paragraphs
```

```python
sentences
```

```python
words = nltk.word_tokenize(corpus)     ### words  -- > sentences
words
```

## Stemming

```python
## Classification problem
## comments of products  --> positive , negative , neutral
## Review ----> Eating, eat, eaten -- eat (stem word) / going , goes ,go ---->  go (stem word)
```

```python
diff_word = ['eat', 'go', 'play', 'running', 'runner', 'easily', 'fairly', 'fairness', 'happily', 'happiness']
```

```python
## Porter Stemmer
from nltk.stem import PorterStemmer
stemming = PorterStemmer()

for word in diff_word:
    print(f"Original word: {word} --> Stemmed word: {stemming.stem(word)}")
```

```python
stemming.stem("congratulations")   ## congratul does not make sense as a word, disadvantage of stemming
```

```python
### Snowball Stemmer
from nltk.stem import SnowballStemmer
```

```python
snow_stemmer = SnowballStemmer("english")
```

```python
for word in diff_word:
    print(f"Original word: {word} --> Snowball Stemmed word: {snow_stemmer.stem(word)}")
```

```python
## Differnce between Porter Stemmer and Snowball Stemmer
stemming.stem("fairly") , stemming.stem("sportingly")  ## porter stemmer
```

```python
snow_stemmer.stem("fairly") , snow_stemmer.stem("sportingly")  ## snowball stemmer
```

## Lemmatization


```python
'''Lemmatization technique is like stemming.
The output we will get is called 'lemma' which is a root word rather than root stem, 
After lemmitization the output will be a meaningful word.
'''
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
lemmatizer.lemmatize("going")
''' 
pos - > noun (n)
pos - > verb (v)
pos - > adjective (a)      

'''
```

```python
for word in diff_word:
    print(f"Original word: {word} --> Snowball Stemmed word: {lemmatizer.lemmatize(word, pos = 'n')}")
```
